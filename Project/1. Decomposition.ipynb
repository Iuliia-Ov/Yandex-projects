## Описание проекта

Пользователи сервиса "Ненужные вещи" совершают много действий в мобильном приложении.  В нем они продают свои ненужные вещи, размещая их на доске объявлений. В имеющихся данных есть инсайты, которые позволят приложению стать лучшим в категориии продажи ненужных вещей. В датасете содержатся данные пользователей, впервые совершивших действия в приложении после 7 октября 2019 года.
Необходимо провести анализ поведения пользователей в мобильном приложении. 

Для этого нам необходимо выполнить следующие задачи:
1. Проанализировать связь целевого события - просмотра контактов - и других действий пользователей.
2. Оценить какие действия чаще совершают те пользователи, которые просматривают контакты.
3. Провести исследовательский анализ данных.
4. Проанализировать влияние событий на совершение целевого события
5. Проверить статистические гипотезы

## Описание данных
В нашем распоряжении два датасета:

[mobile_dataset.csv](https://disk.yandex.ru/d/KaN8Wffo4Q510Q)
 
[mobile_sources.csv](https://disk.yandex.ru/d/RkfgQSEt-zD6Uw)

Файл mobile_dataset.csv ханит информацию о пользователях и совершенными ими действиях. Файл mobile_sources.csv хранит инфомацию об источниках привлечения пользователей.

Структура mobile_sources.csv:
- `userId` — идентификатор пользователя,
- `source` — источник, с которого пользователь установил приложение.

Структура mobile_dataset.csv: 
- `event.time` — время совершения,
- `user.id` — идентификатор пользователя,
- `event.name` — действие пользователя.

Виды действий:
- advert_open — открыл карточки объявления,
- photos_show — просмотрел фотографий в объявлении,
- tips_show — увидел рекомендованные объявления,
- tips_click — кликнул по рекомендованному объявлению,
- contacts_show и show_contacts — посмотрел номер телефона,
- contacts_call — позвонил по номеру из объявления,
- map — открыл карту объявлений,
- search_1—search_7 — разные действия, связанные с поиском по сайту,
- favorites_add — добавил объявление в избранное.

## Содержание проекта

**Цель исследования**

Анализ поведения пользователей и разбор сценариев использования приложения "Ненужные вещи" для дальнейшей сегментации аудитории.

**Загрузка данных и подготовка их к анализу**

Загрузим данные о пользователях приложения, источниках их привлечения и действиях, совершаемх пользователями "Ненужных вещей".

Изучим данные и выполним предобработку. При наличии в данных пропусков и дубликатов выполним определенные действия с ними при неободимости. Убедимся, что типы данных во всех колонках соответствуют сохранённым в них значениям. Обратим внимание на столбцы с датой и временем.

**Исследовательский анализ данных**

1. Составим профили пользователей. 
2. Разделим данные на сессии: примем, что сессия равна календарному дню. 
3. Рассмотрим какие присутсвуют сценарии поведения пользователей, как различается время между распространенными событиями пользователей.
4. Определим какие события влияют на конверсию в целевое действие.
5. Выясним, из каких источников приходят пользователи в приложение. Определим самый популярный источник.
    После каждого пункта сформулируем выводы.

**Основная цель исследования**

Анализ поведения пользователей необходим для управления вовлеченностью аудитории приложения. В ходе анализа нам необходимо будет ответить на следующие вопросы: 
1. Какие события влияют на конверсию в целевое действие?
2. Какие сценарии использования приложения выделяются?
3. Как различается время между распространенными событиями пользователей?

На основе проведенного исследования мы сможем получить гипотезы о том, как  можно улучшить приложение с точки зрения пользовательского опыта.

**Формулировка гипотез**

В представленном датасете имеются данные о пользователях, которые совершают действия *tips_show* и *tips_click*, а также тех, кто совершает только *tips_show*. 

Проверим следующую гипотезу:
конверсия в просмотры контактов различается у этих двух групп.

Кроме того пользователи добавляют в избранное товар (*favorites_add*) после просмотра фотографий (*photos_show*).

Проверим еще одну гипотезу: процент добавления в избранное у объявлений, где был просмотр фотографий товара выше, чем у объявлений без просмотра фотографий.
